```python
#익명함수
def add(x, y):
    return x + y
f = lambda x, y: x + y
print(add(1, 2))
print(f(1,2))
```

```python
def add(x,y):
    return x+y
f=lambda x, y: x+y
print(add(1,2))
print(f(1,2))

f=lambda x:x if x% 3 ==0 else 0
print(f(3))
print(f(4))
```

```python
nums = [1,2,3,4,5,6,7]
def pow(x):
    return x ** 2
f= lambda x: x**2
print(list(map(pow,nums)))
print(list(map(f,nums)))
print(list(map(lambda x: x**2, nums)))
```

```python
ages = [18, 19, 39, 12, 43, 13, 21, 25]
def adult_func(age):
    if age >= 19:
        return True
    else:
        return False
for age in filter(adult_func, ages):
    print(age)
    print()
for age in filter(lambda age: age >= 19, ages):
    print(age)
```
