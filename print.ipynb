```python
age=20
name='홍길동'
str4='이름{}나이{}'.format(name,age)
print(str4)
str5='이름{0}나이{1}'.format(name,age)
print(str5)
str6='이름{1}나이{0}'.format(age,name)
print(str6)
print('이름{1}나이{0}'.format(age,name))
```

```python
age = 20
name='홍길동'
str7='이름{:s} 나이{:d}'.format(name,age)
print(str7)
print('pi={:f}'.format(3.141592))
print('pi={:10f}'.format(3.141592))
print('pi={:5.2f}'.format(3.141592))
print('pi={:.2f}'.format(3.141598))
```
