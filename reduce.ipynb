```python
from functools import reduce #functools를 사용함
nums = [1, 2, 3, 4]
sum = reduce(lambda x, y: x+y, nums)
print(sum)
mul = reduce(lambda x, y: x*y, nums)
print(mul)
```

```python
from functools import reduce
nums = list(range(1,10))
result = reduce(lambda x,y:x*y,filter(lambda num:num%2==0,nums))
print(result)
```

```python
list1 = list(range(1,10))
list2 = list(map(lambda x: x**2, list1))
print(list2)
list3 = [x**2 for x in list1]
print(list3)
list4 = [x**2 for x in range(1,10)]
print(list4)

list5 = [x**2 for x in list1 if x>5 ]
print(list5)

# 조건문에 따라 짝수 값만 필터링
list6 = [x**2 for x in range(1, 10) if x%2==0] #밖에 대괄호 중요
print(list6)
# map 이랑 비슷
```

```python
list1 = list(range(1,11))
list6 = list(range(1,21))
list2 = [x**2 for x in list1]
list3 = [x for x in list6 if x%2==0]
list4 = [x for x in list6 if x%3==0]
list5 = [x**2 for x in list6 if(x%5==0)]
print(list2,list3,list4,list5)
```
