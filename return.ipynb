```python
def devide(num1,num2):
    q= num1 // num2
    r= num1 % num2
    return (q,r)
    
t = devide(25,6)
print(t)
    
```

```python
class Person:
    def __init__(self):
        self.hi = 'hello'
    def hello(self):
        print(self.hi)
person = Person()
person.hello()
print(person.hi)
```

```python
class Person:
    def __init__(self,n,a):
        self.name = n
        self.age= a
    def hello(self):
        print('Hello {} '.format(self.name))
        print('나이는 {} 살입니다'.format(self.age))
person=Person('홍진영',45)
person.hello()
```

```python
class Student:
    def __init__(self,g,b,n):
        self.grade = g
        self.ban = b
        self.name = n
    def introduce(self):
        print('{} 학년 {} 반 {} 입니다'.format(self.grade, self.ban, self.name))
        
student1 = Student(1,4,'신혜빈')
student1.introduce()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        if 0 <= age <= 20: 
            self.__age = age
        else: age = 0
    def inc_age(self):
        self.__age += 1
    def info(self):
        print(self.__age)
person = Person('홍길동',20)
person.inc_age()
person.info()
```

```python
class Person:
    count = 0
    def __init__(self, name):
        self.name = name
        Person.count += 1

person1 = Person('홍길동')
print(person1.count)
person2 = Person('허균')
print(person2.count)
```

```python
# class method static method
# 둘 다 클래스 상태에서 호출
# 일단 static method
# self 가 없음 개체의 속성못씀
class Math:
    @staticmethod
    def add(a,b):
        return a+b
    @staticmethod
    def sub(a,b):
        return a-b
print(Math.add(4,5))
print(Math.sub(9,5))
```

```python
class Rectangle:
    def __init__(self,width,height):
        self.width = width
        self.height = height
    def area(self):
        return self.width*self.height
    
    def double(self):
        self.width *=2
        self.height *=2
        
rect = Rectangle(10,20)
print(rect.area())
rect.double();
print(rect.width,rect.height)
```
