```python
#매개변수가 없는 함수
#define
def hello():
    print('Hello world!!')
    
hello()
```

```python
def hello1(name):
    print('Hello world!!',name)
    
hello1('홍길동')
```

```python
def hello4(greeting, *names):
    for name in names:
        print(greeting, name)

hello4('안녕', 'ㅈ', 'ㅅ','ㅈ','ㄲ','ㅇ')
#  가변 매개변수는 매개변수 중 마지막에만 위치 가능
```

```python
def function(first,second,third):
    return first,second,third
print(function (first=3, second=5, third=7))
print(function(second=5,third=7,first=3))
```

```python
# 매개변수 기본값 = 생략할수있다
def function(first=1,second=3,third=5):
    return first+second+third
print(function(1,2,3))
print(function(1,2))
print(function(1))
print(function(first=1))
print(function(second=5))
print(function(first=1,third=5))
print(function(first=1,second=5))
```

```python
def function1(first,second,third=5):
    return first+second+third

print(function1(1,2,3))
print(function1(1,2))
print(function1(first=1,second=2,third=10))
print(function1(1,second=5))
print(function1(second=5,first=2))
# 언급없는값은 대체 못함
```
